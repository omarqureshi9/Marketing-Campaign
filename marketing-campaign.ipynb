```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv('/kaggle/input/campaign-data/campaign_data.csv')
dataset
```

```python
data=dataset.copy()
data.isnull().sum()
```

```python
sns.scatterplot(x='GoalAmount', y='RaisedAmount', data=data, hue='IsSuccessful')
plt.show()
```

#### This graph shows that a campaign is deemed successful if the raised amount surpasses the goal amount. If the raised amount falls below the goal amount, the campaign is termed "not sucessful"

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='RaisedAmount', y='DurationDays', data=data, hue='IsSuccessful')
plt.show()
```

#### This graph shows that if the raised amount for a campaign is above 100,000, it is sucessful. Additionally, this plot also reflects that if the raised amount is 60,000 or below, the campaign is more likely to be to be unsuccessful than sucessful.
